```python
import pandas as pd
import numpy as np
from costcla import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from costcla.models import CostSensitiveDecisionTreeClassifier
```

```python
#Important notice: Since Costcla is dependant on SCikit learn (joblib) it must be rolled back to 22.2 in order to work (Joblib is removed in 0.23).
#Pip list is mentioned below
```

```python
data = pd.read_csv('Thesis_transformed (1).csv')
```

```python
x = data.drop("ALERT_STATUS", axis=1)
print(x)
print(x.shape)
y = data["ALERT_STATUS"]
print(y)
print(y.shape)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
print(x_train.describe()
,y_train.describe())
```

```python
model = DecisionTreeClassifier(class_weight='balanced', random_state=42)
model.fit(x_train, y_train)
```

```python
optimalModel = DecisionTreeClassifier(min_samples_split=3, max_depth=3, criterion="entropy", ccp_alpha=0.08163265306122448, random_state=42)
optimalModel.fit(x_train, y_train)
```

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=42)
scores = cross_val_score(optimalModel, x, y, scoring="roc_auc_ovo", cv=cv, n_jobs=1)
print("Mean ROC AUC: %.3f" % np.mean(scores))
```

```python
CM= np.array([[0, 0, 0], [5, 55, 55], [50, 50, 50]])
```

```python
CSmodel = CostSensitiveDecisionTreeClassifier()
CSmodel.fit(x_train, y_train, CM).predict(x_test)
```
